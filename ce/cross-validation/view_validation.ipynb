```python
import matplotlib.pyplot as plt
import numpy as np
import pickle
import pandas as pd
```

```python
path1 = "./cv-results/raw-data-cv-results.pkl"
path2 = "./cv-results/rroot-03103-cv-results.pkl"

with open(path1, "rb") as f:
    data1 = pickle.load(f)
with open(path2, "rb") as f:
    data2 = pickle.load(f)

data = {'raw-data':data1['raw-data'], 'r root 0.3103':data2['rroot-03103']}
# to dataframe
df = pd.DataFrame(data)
df.head()
```

```python
# boxplot of raw-data and rroot-03103
plt.style.use('seaborn')
df.boxplot()
plt.ylabel('Accuracy (%)')
# plt.title('Boxplot of raw data and normalized rroot-03103')
plt.show()
```

```python
# boxplot of raw-data and rroot-03103
import seaborn as sns
ax = sns.boxplot(data=df)
plt.ylabel('Accuracy (%)')
# plt.title('Boxplot of raw data and normalized rroot-03103')
# save plot
plt.savefig('cross-validation.pdf', format='pdf')
plt.show()
```
